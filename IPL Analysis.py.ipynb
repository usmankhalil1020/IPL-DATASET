```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl = pd.read_csv(r"E:/USMAN DATA/Datasets WScubetech/Sir Umaid Datasets/matches.csv")
print(ipl)
```

```python
ipl.shape
```

```python
ipl.head()
```

```python
ipl.columns
```

```python
#Getting frequency of most man of the matches
ipl["player_of_match"].value_counts()
```

```python
ipl["player_of_match"].value_counts()[0:10]
```

```python
list(ipl["player_of_match"].value_counts()[0:5].keys())
```

```python
sns.set(rc = {"figure.figsize":(8,5)})
plt.xlabel("Players")
plt.ylabel("Total man of the matches")
plt.bar(list(ipl["player_of_match"].value_counts()[0:5].keys()),list(ipl["player_of_match"].value_counts()[0:5]),color=["r","b","y","g","b"],)
plt.show()
```

```python
#Getting the frequency of result columns
ipl["result"].value_counts()
```

```python
#Find out no of toss wins w.r.t each teams
ipl["toss_winner"].value_counts()
```

```python
#Extracting a record where a team won batting first
batting_first=ipl[ipl["win_by_runs"]!=0]
```

```python
batting_first.head()
```

```python
plt.hist(batting_first["win_by_runs"])
plt.xlabel("Runs")
plt.ylabel("Matches")
plt.title("Distribution of Runs")
sns.set(rc = {"figure.figsize":(5,5)})
plt.show()
```

```python
#Finding out number of wins w.r.t each team after batting first
batting_first["winner"].value_counts()
```

```python
#Working a bar plot for top 3 teams with most wins after batting first
sns.set(rc = {"figure.figsize":(6,6)})
plt.xlabel("Teams")
plt.ylabel("Number of Matches")
plt.bar(list(batting_first["winner"].value_counts()[0:3].keys()),list(batting_first["winner"].value_counts()[0:3]),color=["blue","yellow","orange"])
plt.show()
```

```python
sns.set(rc = {"figure.figsize":(6,6)})
plt.pie(list(batting_first["winner"].value_counts()),labels=list(batting_first["winner"].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```

```python
#Extracting those records where a team won after batting second
batting_second=ipl[ipl["win_by_wickets"]!=0]
```

```python
batting_second.head()
```

```python
#Woring a histogram for frequency of wins w.r.t no of wickets
plt.hist(batting_second["win_by_wickets"], bins=10)
plt.xlabel("wickets")
plt.ylabel("Matches")
plt.title("Win by wickets")
sns.set(rc = {"figure.figsize":(7,4)})
plt.show()
```

```python
#Find out the frequency of number of wins w.r.t each time after batting second
batting_second["winner"].value_counts()
```

```python
#Working a bar-plot for top 3 teams with most wins after batting second
plt.bar(list(batting_second["winner"].value_counts()[0:3].keys()),list(batting_second["winner"].value_counts()[0:3]),color=["purple","blue","red"])
sns.set(rc = {"figure.figsize":(7,5)})
plt.xlabel("Teams")
plt.ylabel("No of Matches")
plt.show()
```

```python
#Working a pie chart for distribution of most wins after batting second
sns.set(rc = {"figure.figsize":(6,6)})
plt.pie(list(batting_second["winner"].value_counts()),labels=list(batting_second["winner"].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```
